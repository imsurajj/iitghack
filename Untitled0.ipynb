```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Upload your dataset to Colab first (adjust the path if needed)
data = pd.read_csv ('/content/dataset.csv')
print("Columns:", data.columns)
data.head()

# Combine date and time columns to create a 'Time' column
data['Time'] = pd.to_datetime(data['LastUpdatedDate'] + ' ' + data['LastUpdatedTime'], format="%d-%m-%Y %H:%M:%S")

# Map traffic condition to numerical weights
traffic_map = {'low': 1.0, 'average': 2.0, 'high': 3.0}
data['TrafficConditionNearby'] = data['TrafficConditionNearby'].map(traffic_map)

print("Number of lots:", data['SystemCodeNumber'].nunique())
print("Number of time steps:", data['Time'].nunique())
print(data.describe())
# Map vehicle types to numerical weights
vehicle_map = {'car': 1.0, 'bike': 0.5, 'truck': 1.5}
data['VehicleTypeWeight'] = data['VehicleType'].map(vehicle_map)

# Set base price
BASE_PRICE = 10

# Unique IDs and time steps
lots = data['SystemCodeNumber'].unique()
# Use the unique datetime values as time steps
time_steps = data['Time'].unique()


# Initialize price history dictionaries for all 3 models
price_history_model1 = {lot: [BASE_PRICE] for lot in lots}
price_history_model2 = {lot: [BASE_PRICE] for lot in lots}
price_history_model3 = {lot: [BASE_PRICE] for lot in lots}
alpha = 0.5

def model_1(prev_price, occupancy, capacity):
    return prev_price + alpha * (occupancy / capacity)
# Demand function weights
beta, gamma, delta, epsilon = 0.4, 0.2, 0.3, 0.1
lambda_param = 0.3

def compute_demand(row):
    return (
        beta * (row['Occupancy'] / row['Capacity']) +
        gamma * row['QueueLength'] -
        delta * row['TrafficConditionNearby'] +
        epsilon * row['IsSpecialDay'] +
        row['VehicleTypeWeight']
    )

def normalize_demand(demand_series):
    return (demand_series - demand_series.min()) / (demand_series.max() - demand_series.min() + 1e-6)

def model_2(base_price, normalized_demand):
    price = base_price * (1 + lambda_param * normalized_demand)
    return np.clip(price, 0.5 * base_price, 2.0 * base_price)
from geopy.distance import geodesic

# Compute distances between all lot pairs
lot_locations = data.groupby('SystemCodeNumber')[['Latitude', 'Longitude']].first()
dist_matrix = pd.DataFrame(index=lots, columns=lots, dtype=float)

for lot1 in lots:
    for lot2 in lots:
        coord1 = (lot_locations.loc[lot1, 'Latitude'], lot_locations.loc[lot1, 'Longitude'])
        coord2 = (lot_locations.loc[lot2, 'Latitude'], lot_locations.loc[lot2, 'Longitude'])
        dist_matrix.loc[lot1, lot2] = geodesic(coord1, coord2).meters

def get_competitor_prices(lot_id, history, radius=500):
    competitors = dist_matrix.loc[lot_id][dist_matrix.loc[lot_id] < radius].index
    competitors = competitors.drop(lot_id, errors='ignore')
    return [history[c][-1] for c in competitors if len(history[c]) > 0]

# Sort data by Time to process chronologically
data = data.sort_values(by='Time')


for t in time_steps:
    df_t = data[data['Time'] == t]
    demands = []

    # Compute demand for Model 2
    for _, row in df_t.iterrows():
        demand = compute_demand(row)
        demands.append(demand)
    norm_demands = normalize_demand(pd.Series(demands))

    # Update prices for each lot
    for i, (_, row) in enumerate(df_t.iterrows()):
        lot = row['SystemCodeNumber']

        # Previous prices
        prev_price_1 = price_history_model1[lot][-1]
        prev_price_2 = price_history_model2[lot][-1]
        prev_price_3 = price_history_model3[lot][-1]

        # Model 1
        new_price_1 = model_1(prev_price_1, row['Occupancy'], row['Capacity'])
        price_history_model1[lot].append(new_price_1)

        # Model 2
        new_price_2 = model_2(BASE_PRICE, norm_demands.iloc[i])
        price_history_model2[lot].append(new_price_2)

        # Model 3 (Competitive)
        competitors = get_competitor_prices(lot, price_history_model3)
        if competitors:
            avg_comp_price = np.mean(competitors)
            if row['Occupancy'] >= row['Capacity']:
                adjustment = -0.1 * (avg_comp_price - BASE_PRICE)
            else:
                adjustment = 0.1 * (avg_comp_price - BASE_PRICE)
        else:
            adjustment = 0

        price_model_3 = new_price_2 + adjustment
        price_model_3 = np.clip(price_model_3, 0.5 * BASE_PRICE, 2.0 * BASE_PRICE)
        price_history_model3[lot].append(price_model_3)
plt.figure(figsize=(18, 8))

for lot in lots:
    # Ensure price history has more than one point to plot a line
    if len(price_history_model1[lot]) > 1:
        plt.plot(price_history_model1[lot], label=f'Lot {lot} - Model 1', linestyle='--')
    if len(price_history_model2[lot]) > 1:
         plt.plot(price_history_model2[lot], label=f'Lot {lot} - Model 2')
    if len(price_history_model3[lot]) > 1:
        plt.plot(price_history_model3[lot], label=f'Lot {lot} - Model 3', linestyle=':')


plt.title('Dynamic Pricing Over Time for All Parking Lots')
plt.xlabel('Time Steps')
plt.ylabel('Price ($)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

This notebook implements three dynamic pricing models for urban parking lots:

- Model 1: Baseline Linear Pricing
- Model 2: Demand-Based Pricing
- Model 3: Competitive Pricing

Libraries used: NumPy, Pandas, Matplotlib.
## Model 1: Baseline Linear Pricing
Price increases linearly with occupancy ratio:

**Formula:**
Price_t+1 = Price_t + α * (Occupancy / Capacity)
## Model 2: Demand-Based Pricing
Price depends on a demand function of:

- Occupancy rate
- Queue length
- Traffic level
- Special day
- Vehicle type weight

Demand is normalized to ensure smooth price variation.
## Model 3: Competitive Pricing
- Considers competitor prices based on geographic proximity.
- Adjusts own price if:
    - Lot is full and nearby lots are cheaper (reduce price).
    - Nearby lots are expensive (increase price).
## Conclusion

We implemented three dynamic pricing models:

- **Model 1:** Simple linear baseline responding only to occupancy.
- **Model 2:** Demand-based, considering multiple features.
- **Model 3:** Adds competitive pricing via spatial analysis.

These models demonstrate increasing sophistication in real-time pricing strategies. You can extend this work by:
